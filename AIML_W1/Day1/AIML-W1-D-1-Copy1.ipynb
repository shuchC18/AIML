### Q9. Write a python program that can calculate the `mean` and `median` of following list:
### `data = [1,1,2,4,2,6,9,5,3,7,3,8]`

````python
data = [1,1,2,4,2,6,9,5,3,7,3,8]

i1 = 0

for j1 in data:
    i1 += j1
print(i1)
````

````python
#Mean

mean1 = i1 / len(data)

mean1
````

````python
data.sort()

data
````

````python
#Median 

n = len(data)

if n % 2 == 0:
    median = (data[(n//2)-1] + data[n//2]) // 2
        
else:
    median = data[n//2]
        
print("Median =", median)
````

````python
#Variance
mean1 = 4.25

var = 0

n = len(data) - 1

for x in data:
    var += ((x - mean1)**2) / n
print("Variance =", var)
````

````python
#Standard Deviation

SD = (var)**(1/2)

SD
````
