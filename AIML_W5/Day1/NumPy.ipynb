Ways to Represent NumPy arrays

1. np.array(): array of numbers
2. np.arange(): numbers in the specified range
3. np.linspace(): numbers in given range with specifed number of data points
4. np.logspace(): specified logarithmic numbers within the specifed range
5. np.zeros(): specified number of zeros
6. np.ones(): specified number of ones
7. np.identity(): specifed number of arrays with one of the sub-element unity
8. np.diag(): specifed elemnts are presented diagoally
